```python
# Create new column named "Annual Skiier Visits"
df["Annual Skiier Visits"] = ""

# Rename column as "Ski Resort"
df = df.rename(columns={0:"Ski Resort"})
```

```python
df
```

```python
# Delete unnecessary rows
ski_df =df.drop([0, 1, 2, 3, 6, 9, 12, 15, 18, 21, 24, 27, 30, 33, 34])
```

```python
ski_df
```

```python
# Moved every other row to a new column labled "Annual Skiier Visits"
# https://stackoverflow.com/questions/36181622/moving-every-other-row-to-a-new-column-and-group-pandas-python

visits_df = pd.DataFrame({'Ski Resort':ski_df['Ski Resort'].iloc[::2].values, 'Annual Skiier Visits':ski_df['Ski Resort'].iloc[1::2].values})
visits_df
```

```python
# Sliced 'Ski Resort' names to remove rank # and place in new column titled "Resort" 
visits_df['Rank'], visits_df['Resort'] = map(visits_df["Ski Resort"].str.slice, [0, 8]) 
visits_df
```

```python
# Split location of resort in 'Resort' column and placed clean name of ski resort in new column titled 'Resort Name'

ski_visits_df = visits_df["Resort"].str.split(",", n = 1, expand = True)
visits_df["Resort_Name"] = ski_visits_df[0]
visits_df["State"] = ski_visits_df[1]
visits_df
```

```python
# Split Annual Skiier Visits to remove "Annual Skiier Visits" and input solely the number of skiiers in new column titled "Number"

ski_visits_df = visits_df["Annual Skiier Visits"].str.split(":", n = 1, expand = True)
visits_df["Annual"] = ski_visits_df[0]
visits_df["Number"] = ski_visits_df[1]
visits_df
```

```python
# Deleting unnecessary columns titled ['Ski Resort', 'Annual Skiier Visits', 'Rank', 'Resort', 'Annual'] from visits_df

                                                     
visits_df = visits_df.drop(columns=['Ski Resort', 'Annual Skiier Visits', 'Rank', 'Resort', 'Annual'])                                                    
visits_df
```
